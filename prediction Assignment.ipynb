```python
# 2. Consider a dataset with a categorical feature "Country" having 10 unique values. 
# Having other Columns state,Population,Density,Area. Make predictions.

import pandas as pd
import numpy as np
data = {
    'Country': ['USA', 'Canada', 'India', 'Germany', 'China', 'Brazil', 'Australia', 'Japan', 'South Africa', 'Mexico'],
    'State': ['CA', 'ON', 'MH', 'BE', 'BJ', 'BR', 'NSW', 'TK', 'GP', 'CMX'],
    'Population': [331002651, 37742154, 1380004385, 83783942, 1444216107, 212559417, 25499884, 126476461, 59308690, 128932753],
    'Density': [36.77, 4.08, 464.24, 237.97, 153.00, 25.04, 3.25, 349.78, 47.42, 66.44],
    'Area': [9147590, 9976140, 2973190, 357022, 9596960, 8515767, 7692024, 377975, 1221037, 1964375],
}
df = pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
#to standardise the values
from sklearn.preprocessing import MinMaxScaler

column_to_scale = ['Population','Density','Area']
scaler = MinMaxScaler()
df[column_to_scale] = scaler.fit_transform(df[column_to_scale])
df
```

```python
#to train the dataset
from sklearn.model_selection import train_test_split
X = df[['Population','Area']]
y = df['Density']

X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=42)
```

```python
print("X train:\n",X_train, 
      "\nX test:\n", X_test,
      "\ny train:\n", y_train,
      "\ny test:\n",y_test)
```

```python
# to make prediction
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
print("Predicted values :\n", y_pred)
```
